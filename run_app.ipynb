# Configure Python Path

```python
import os
import sys

# Automatically locate project root (folder that contains habit_tracker/)
current_path = os.getcwd()

# If notebook is inside habit_tracker/, move one level up
if current_path.endswith("habit_tracker"):
    project_root = os.path.abspath("..")
else:
    project_root = current_path

# Add project root to PYTHONPATH
if project_root not in sys.path:
    sys.path.insert(0, project_root)

print("Project root set to:", project_root)
print("Imports will now work.")
```

## Import HabitManager

```python
from habit_tracker.manager import HabitManager

m = HabitManager()
m
```

## Create Sample Habits

```python
m.create_habit(
    name="Log one Future-Self win",
    description="Daily reflection to stay aligned with long-term goals.",
    periodicity="daily"
)

m.create_habit(
    name="Document one learning",
    description="Capture a new insight or skill every day.",
    periodicity="daily"
)

m.create_habit(
    name="Nurture a relationship",
    description="Check in with someone weekly.",
    periodicity="weekly"
)

print("Habits created successfully.")
```
